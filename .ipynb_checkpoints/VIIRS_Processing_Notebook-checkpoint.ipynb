# Processing and interacting with NASA's Visible Infrared Imaging Radiometer Suite (VIIRS) Surface Reflectance products

The Land Processes Distributed Active Archive Center (LP DAAC) distributes NASA Visible Infrared Imaging Radiometer Suite (VIIRS) Surface Reflectance products. The NASA VIIRS Surface Reflectance products are archived and distributed in a file format call HDF-EOS5. The Hierarchical Data Format version 5 (HDF5) is a NASA selected format of choice for standard science product archival and distribution, and is the underlying format for HDF-EOS5. HDF-EOS is the standard format and I/O library for the Earth Observing System (EOS) Data and Information System (EOSDIS). HDF-EOS5 extends the capabilities of the HDF5 storage format, adding support for the various EOS data types (e.g., Point, Swath, and Grid) into the HDF5 framework.

This tutorial demonstrates how to open and interact with VIIRS Surface Reflectance data files (HDF-EOS5 (.h5)) in a python environment. Additionally, this tutorial demonstrates how to define the coordinate reference system (CRS) and export each science dataset layer as a GeoTIFF file that can be loaded with spatial referencing into a GIS and/or Remote Sensing software program. Visit the [LP DAAC website](https://lpdaac.usgs.gov/product_search/?keyword=Surface+Reflectance&query=VIIRS&sort=title&view=cards) for more information about NASA VIIRS Surface Reflectance products.

# 1. Getting Started
## 1a. Import Packages
### Import the python modules you'll need to in order to interact with the HDF-EOS5 (.h5) file.

```python
import h5py
import numpy as np
from osgeo import gdal, gdal_array
import os
```

***
## 1b. Set Up the Working Environment
#### Define the input directory containing the VNP09A1.001 files, and create an output directory to store a series of GeoTIFFs that will be created from this tutorial.

```python
inDir = input("Enter input directory: ")                                # enter the dir on your OS containing VIIRS files
os.chdir(inDir)                                                         # Change to working directory
outDir = os.path.normpath(inDir + os.sep + 'output') + '\\'                   # Set output directory
if not os.path.exists(outDir): os.makedirs(outDir)                      # Create output directory
```

***
## 1c. Retrieve Files
### Identify the VIIRS .h5 file to process and interact with.
#### In this example, we are describing a situation where a series of VIIRS .h5 files have been downloaded to a single folder in a directory. The following will find all of the .h5 files in an input directory and create a python list of all the file names.
#### Make sure that you download the files into the __inDir__ directory defined above to follow along in the tutorial.


```python
fileList = [file for file in os.listdir(inDir) if file.endswith('.h5')]    # Read all the .h5 files within the input directory
for f in fileList:                                                         # Print the .h5 file names
    print(f)
```

#### Select a file from the list to process and interact with.

```python
inFile = fileList[0]    # NOTE: fileList[0] selects the 1st file in the list
inFile
```

***
## 1d. Specify the filename for the output GeoTIFFs that will be generated later

```python
inFileLoc = '{}/{}'.format(inDir, inFile)      # Combine directory to .h5 file
fileName = inFileLoc.split('/')[-1][:-3]       # Specify the filename
fileName                                       # print the filename
```

***
<a id="section2"></a>
# 2. Importing Data
## 2a. Read in a VIIRS HDF-EOS5 file

```python
f = h5py.File('{}'.format(inFileLoc), "r")         # Read in VIIRS HDF-EOS5 file and set to variable
```

***
## 2b. Navigate the HDF-EOS5 file
#### The VIIRS HDF-EOS5 file contains two groups in which data and metadata is stored.

```python
list(f)        # List groups in file
```

#### The _HDFEOS_ group is where you access the VIIRS data. 

```python
list(f['HDFEOS'])   # List contents of HDFEOS group
```

#### The _HDFEOS INFORMATION_ group contains the metadata for the file. The _StructMetadata.0_ object within stores the metadata information.

```python
list(f['HDFEOS INFORMATION'])     # List contents of HDFEOS Information group
```

#### All of the objects in the VIIRS HDF-EOS5 file can be identified using the following commands:

```python
all_h5_objs = []              # Create empty list
f.visit(all_h5_objs.append)   # Walk through directory tree, retrieve objects and append to list
all_h5_objs
```

***
<a id="section3"></a>
# 3. Georeferencing

#### Get the geoinformation from the _StructMetadata.0_ object within the HDF-EOS5 file.

```python
f_metadata = f['HDFEOS INFORMATION']['StructMetadata.0'][()].split()   # Read file metadata
f_metadata_byte2str = [s.decode('utf-8') for s in f_metadata]          # Clean up file metadata
f_metadata_byte2str                                                   # Print the file metadata record
```

#### Identify the upper left longitude and latitude coordinates from the metadata.

```python
ulc = [i for i in f_metadata_byte2str if 'UpperLeftPointMtrs' in i]                      # Search file metadata for the upper left corner of the file
ulcLon = float(ulc[0].replace('=', ',').replace('(', '').replace(')', '').split(',')[1]) # Parse metadata string for upper left corner lon value
ulcLat = float(ulc[0].replace('=', ',').replace('(', '').replace(')', '').split(',')[2]) # Parse metadata string for upper left corner lat value
```

#### VIIRS climate model grids (CMG) are mapped to a different coordinate reference system than the other VIIRS level 3 products. The VIIRS CMG datasets are referenced to the Geographic coordinate space, while the other VIIRS level 3 products are referenced to a projected Sinusoidal coordinate system identical to the one used for the Moderate Resolution Imaging Spectroradiometer (MODIS) collection.

```python
projInfo = {'SINU':'PROJCS["unnamed",GEOGCS["Unknown datum based upon the custom spheroid", DATUM["Not specified (based on custom spheroid)", SPHEROID["Custom spheroid",6371007.181,0]],PRIMEM["Greenwich",0], UNIT["degree",0.0174532925199433]], PROJECTION["Sinusoidal"],PARAMETER["longitude_of_center",0],PARAMETER["false_easting",0],PARAMETER["false_northing",0],UNIT["Meter",1]]',
            'GEO':'GEOGCS["Unknown datum based upon the Clarke 1866 ellipsoid", DATUM["Not specified (based on Clarke 1866 spheroid)", SPHEROID["Clarke 1866",6378206.4,294.9786982139006]], PRIMEM["Greenwich",0], UNIT["degree",0.0174532925199433]]'}
```

#### Extract the _GridName_ element from the metadata. This will be used to determine if the file we're working with is a CMG or one of the other VIIRS level 3 products.

```python
gridName = [i.split('=')[1].replace('\"', '') for i in f_metadata_byte2str if 'GridName=' in i] # Extract the GridName from the metadata
gridName                                                                                        # Print GridName
```

#### Assign the projection parameters based on the _gridName_ python object

```python
if gridName == 'VNP_CMG':         # Determine if the data is VNP09CMG or not  
    prj = projInfo['GEO']
else:
    prj = projInfo['SINU']
```

***
<a id="section4"></a>
# 4. Identify the Science Datasets (SDS) in the HDF-EOS5 file and SDS Metadata

#### Generate a list of all the SDS layers

```python
all_datasets = [obj for grid in gridName for obj in all_h5_objs if isinstance(f[obj],h5py.Dataset) and grid in obj]

all_datasets
```

#### Extract the red (SurfReflect_M5), green (SurfReflect_M4), and blue (SurfReflect_M3) SDS layers from the HDF-EOS5 file

```python
rObj = f['{}'.format(all_datasets[10])]      # M5 = Red
gObj = f['{}'.format(all_datasets[9])]       # M4 = Green
bObj = f['{}'.format(all_datasets[8])]       # M3 = Blue
```

#### List the attributes for the _rObj_ and extract the scale factor and fill value

```python
list(rObj.attrs)       # list of Red layer attributes
```

```python
scaleFactor = rObj.attrs['scale_factor'][0]   # Set scale factor to a variable
fillValue = rObj.attrs['_FillValue'][0]       # Set fill value to a variable
```

***
<a id="section5"></a>
# 5. Visualize and Process VIIRS SDS - VNP09A1 Example

#### Use matplotlib to plot images using VIIRS SurfReflect_M5 (R), SurfReflect_M4 (G), SurfReflect_M3 (B), and SurfReflect_M7 (NIR) layers.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

***
## 5a. Extract the data for each SDS and apply the scale factor
#### The scale factor is the same for all surface reflectance M-bands, so you can simply use the scale factor from the red band for the other bands.

```python
rArr = rObj[()] * scaleFactor   
gArr = gObj[()] * scaleFactor
bArr = bObj[()] * scaleFactor
```

***
## 5b. Create an RGB array

```python
rgbArray = np.dstack((rArr,gArr,bArr))   # Create RGB array stack
```

***
## 5c. Specify the fill values

```python
rgbArray[rgbArray == fillValue * scaleFactor] = np.nan
```

***
## 5d. Visualize the RGB image

```python
plt.figure(figsize=(10,10))                               # Set the figure size
plt.axis('off')                                           # Remove axes from plot
plt.imshow(rgbArray, interpolation='nearest', alpha=0.9)  # Display RGB as an image 
```

***
## 5e. Calculate the Normalized Difference Vegetation Index (NDVI)
#### Calculate NDVI using the NIR (SurfReflect_M7) and red (SurfReflect_M5) layers

```python
nirArr = f['{}'.format(all_datasets[11])][()] * scaleFactor   # Apply scale factor to Near Infrared (M7) band
nirArr[nirArr == fillValue * scaleFactor] = np.nan            # Specify the fill value for NIR band
rArr[rArr == fillValue * scaleFactor] = np.nan                # Specify fill value for the red band
```

```python
ndvi = (nirArr - rArr) / (nirArr + rArr)                      # Calculate NDVI
```

***
## 5f. Visualize the NDVI array

```python
plt.figure(figsize=(10,10))        # Set the figure size
plt.axis('off')                    # Remove axes from plot
plt.imshow(ndvi, cmap = 'YlGn')    # Display NDVI 
```

#### Compute histogram to better stretch the contrast of the image

```python
plt.hist(ndvi, bins=50);                  # Plot a histogram
plt.xticks(np.arange(-1.0, 1.25, 0.25));  # Set the x ticks
```

**Note:** The semi-colon at the end suppresses the output array that prints with the histogram.

```python
import warnings
warnings.filterwarnings("ignore")

fig = plt.figure(figsize=(10,10))                                                                     # Set the figure size
fig.suptitle('VIIRS Normalized Difference Vegetation Index (NDVI)', fontsize=12, fontweight='bold')   # Add subtitle to plot
plt.axis('off')                                                                                       # Remove axes from plot
ax = fig.add_subplot(111)                                                                             # Create a subplot
fig.subplots_adjust(top=0.93)                                                                         # tune the bottom of the subplots
ax.set_title('Tile: h09v05')                                                                          # Add title to plot
ax1 = plt.gca()                                                                                       # Get a reference to the current axes
im = ax1.imshow(ndvi, vmin = 0, vmax = 0.9, cmap = 'YlGn')                                            # Specify the min and max in display of NDVI value
plt.colorbar(im, orientation='horizontal', fraction=0.047, pad=0.02, label = 'NDVI').outline.set_visible(False) # Add Color bar
fig.savefig('{}/{}_ndvi.png'.format(outDir, fileName))                                                # Save figure in output directory
```

# 6. Create GeoTIFF outputs

#### Loop through each SDS in the VIIRS HDF-EOS5 file. Create a georeferenced GeoTIFF for each SDS.

```python
for ds in all_datasets:
    dsName = ds.split('/')[-1]
    print('Processing: {}'.format(dsName))
    dsArray = f[ds][()]
    nRow = dsArray.shape[0]
    nCol = dsArray.shape[1]
    '''Currently, VIIRS HDF-EOS5 files do not contain information regarding the spatial resolution of the dataset within.'''
    if nRow == 1200:                      # VIIRS A1 - 1km or 1000m
        yRes = -926.6254330555555    
        xRes = 926.6254330555555
    elif nRow == 2400:                    # VIIRS H1 - 500m
        yRes = -463.31271652777775
        xRes = 463.31271652777775
    elif nRow == 3600 and nCol == 7200:    # VIIRS CMG
        yRes = -0.05
        xRes = 0.05
        # Re-set upper left dims for CMG product                
        ulcLon = -180.00
        ulcLat = 90.00
    geoInfo = (ulcLon, xRes, 0, ulcLat, 0, yRes)
    
    dataType = gdal_array.NumericTypeCodeToGDALTypeCode(dsArray.dtype)
    format = "GTiff"
    driver = gdal.GetDriverByName(format)
    print('Creating {}_{}.tif'.format(fileName, dsName))
    out_ds = driver.Create('{}/{}_{}.tif'.format(outDir, fileName, dsName), nCol, nRow, 1, dataType)
    out_ds.GetRasterBand(1).WriteArray(dsArray)
    out_ds.SetGeoTransform(geoInfo)
    '''The projection parameters were specified and assigned earlier in the tutorial. See proj object.'''
    out_ds.SetProjection(prj)
    out_ds = None
    print('Output file: {}_{}.tif'.format(fileName, dsName))
    
```
